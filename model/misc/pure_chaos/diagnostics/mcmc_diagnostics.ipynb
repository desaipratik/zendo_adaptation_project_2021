```python
%matplotlib inline
import numpy as np
import scipy.optimize as opt

def choice(tau,raw_probs,data):
    probs = []
    for a,b in zip(raw_probs, data):
        prob = math.e**(a/tau) / (math.e**(a/tau) + math.e**((1-a)/tau))
        if b == 0:
            prob = 1 - prob
        probs.append(prob)
    return -np.sum([np.log(i) for i in probs])


results = opt.minimize(choice,1, args=([0.1,0.1],[True,True]))
print("Solution: x=%f" % results.x[0])
```
