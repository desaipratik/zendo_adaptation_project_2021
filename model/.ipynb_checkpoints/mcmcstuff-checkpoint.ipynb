```python
import numpy as np
import random
from scipy.optimize import minimize
raw_probs = [0.8200000000000001, 0.8200000000000001, 0.8200000000000001, 0.8200000000000001, 0.8200000000000001, 0.8200000000000001, 0.8200000000000001, 0.8200000000000001]
data = [1,1,1,1,0,0,0,0]

tau = 1
def choice(tau, data, raw_probs):
    probs = []
    for i in raw_probs:
        prob = math.e**(i/tau) / (math.e**(i/tau) + math.e**((1-i)/tau))
        if data[raw_probs.index(i)] == 0:

            prob = 1 - prob
       
            
        probs.append(prob)
    return -np.sum([np.log(i) for i in probs])


print(minimize(choice, 1, args=(raw_probs,data)))
print(choice(tau, data, raw_probs))
```

```python
%matplotlib inline
import numpy as np
import scipy.optimize as opt

def choice(tau,raw_probs,data):
    probs = []
    for a,b in zip(raw_probs, data):
        prob = math.e**(a/tau) / (math.e**(a/tau) + math.e**((1-a)/tau))
        if b == 0:
            prob = 1 - prob
        probs.append(prob)
    return -np.sum([np.log(i) for i in probs])


results = opt.minimize(choice,1, args=([0.1,0.1],[True,True]))
print("Solution: x=%f" % results.x[0])
```
